```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv(r'C:\Users\Khazir\Desktop\movies.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df_features = ['genres','keywords','tagline','cast','director']
print(df_features)
```

```python
for i in df_features:
    df[i] = df[i].fillna('')
```

```python
combined_features = df['genres']+' '+df['keywords']+' '+df['tagline']+' ' +df['cast']+' '+df['director']
```

```python
print(combined_features)
```

```python
vectorizer = TfidfVectorizer()
```

```python
features_vectors = vectorizer.fit_transform(combined_features)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(features_vectors)
```

```python
print(similarity.shape)
```

```python
movie = input("Enter the name of your favourite movie:")
```

```python
my_list = df['title'].tolist()
```

```python
find_close_match = difflib.get_close_matches(movie,my_list)
```

```python
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
index_of_the_movie = df[df.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score,key = lambda x:x[1] , reverse = True)
print(sorted_similar_movies)
```

```python
print("Movies suggested for you \n")

i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.index == index]["title"].values[0]  
  if(i<11):
    print(i,'.',title_from_index)
    i+=1
```
